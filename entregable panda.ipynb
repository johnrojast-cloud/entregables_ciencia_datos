```python
# panda aka pd
import pandas as pd
```

```python
# import json
url = "https://www.datos.gov.co/resource/a77s-vxqq.json"
df_json = pd.read_json(url)
print(df_json)
```

```python
# 5 primeros
print(df_json.head())
```

```python
# 5 ultimos
print(df_json.tail())
```

```python
# data frame info
print(df_json.info())
```

```python
# basics stadistics
print(df_json.describe())
```

```python
# columns name
print(df_json.columns)
```

```python
# index
print(df_json.index)
```

```python
# column expedientes
print(df_json["expediente"])
```

```python
# columns pais y ciudad
print(df_json[["pais","ciudad"]])
```

```python
# file 50
print(df_json.iloc[50])
```

```python
# files and columns
print(df_json.iloc[0:3, 0:3])
```

```python
# order
order = df_json.sort_values("expediente")
print(order)
```

```python
# descend order
order_desc = df_json.sort_values("expediente", ascending=False)
print(order_desc)
```

```python
 # delete direccion column
 df = df_json.drop("direccion", axis=1)
print(df)
```

```python
# delete file 1
df = df_json.drop(1, axis=0)
print(df)
```

```python
# save in csv
df_json.to_csv("data.csv", index=False)
```

```python
# open csv
new_df = pd.read_csv("data.csv")
print(new_df)
```
